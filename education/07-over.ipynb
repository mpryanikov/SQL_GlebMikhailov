### последняя дата активности каждого пользователя:

***rank():***

```python
sql = '''select t.*,
rank() over (partition by t.user_id order by t.dt desc) as rnk
from revenue t
'''
```

```python
select(sql)
```

```python
sql = '''with 
dt_rank as (
    select t.*,
    rank() over (partition by t.user_id order by t.dt desc) as rnk
    from revenue t
)
select * from dt_rank t
where t.rnk = 1
'''
```

```python
select(sql)
```

***row_number():***

```python
sql = '''select t.*,
row_number() over (partition by t.user_id order by t.dt desc) as rnk
from revenue t
'''
```

```python
select(sql)
```

```python
sql = '''with 
dt_rank as (
    select t.*,
    row_number() over (partition by t.user_id order by t.dt desc) as rnk
    from revenue t
)
select * from dt_rank t
where t.rnk = 1
'''
```

```python
select(sql)
```

## 3. Топ 3 зарплаты в отделе (задача на интервью)

```python
t = pd.DataFrame({'dep':['a','a','a','a','a',
                         'b','b','b','b','b'],
                  'emp':['aa','bb','cc','dd','ee',
                         'aa','bb','cc','dd','ee'],
                  'sal':[5,5,3,2,1,
                         5,4,3,2,1]})
```

```python
cur = conn.cursor()
sql = '''
drop table if exists salary;
CREATE TABLE salary (
    dep       varchar(max),
    emp       varchar(max),
    sal       int
);
'''
cur.execute(sql)
conn.commit()
for index,row in t.iterrows():
    cur.execute('''INSERT INTO salary(
                    [dep],[emp],[sal]
                    ) 
                    values (?,?,?)
    ''', 
                    row['dep'],
                    row['emp'], 
                    row['sal']
               )
conn.commit()
cur.close()
sql = '''select * from salary t'''
select(sql)
```

```python
sql = '''select t.*,
rank() over (partition by t.dep order by t.sal desc) as rnk_rank,
dense_rank() over (partition by t.dep order by t.sal desc) as rnk
from salary t
'''
```

```python
select(sql)
```

```python
sql = '''with 
salary_rnk as (
    select t.*,
    dense_rank() over (partition by t.dep order by t.sal desc) as rnk
    from salary t
)
select * from salary_rnk t
where t.rnk <= 3
'''
```

```python
select(sql)
```
